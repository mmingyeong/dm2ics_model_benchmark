### 3. 유틸 정의: 배치 정리, 손실 분해, 로그/시각화 헬퍼

- `prepare_batch`: 채널/디바이스 정리  
- `compute_gan_loss_components`: loss breakdown (이미 정의돼 있다면 재사용)  
- `log_step`: TensorBoard/내부 로그 기록  
- `plot_gan_diagnostics`: 추적한 기록을 1x3 시각화


```python
# prepare_batch (디바이스/채널 맞추기)
def prepare_batch(x, y, model):
    device = next(model.parameters()).device
    if x.ndim == 4:
        x = x[:, None]
    if y.ndim == 4:
        y = y[:, None]
    x = x.float().to(device)
    y = y.float().to(device)
    return x, y

# diagnostics plotting (재정의 안된 경우)
def plot_gan_diagnostics(df, smooth_window=5):
    def smooth(x):
        return x.rolling(smooth_window, min_periods=1, center=True).mean()
    fig, axs = plt.subplots(1, 3, figsize=(18, 4), constrained_layout=True)
    axs[0].plot(df['step'], smooth(df['loss_G']), label='Generator Loss')
    axs[0].plot(df['step'], smooth(df['loss_D']), label='Discriminator Loss')
    axs[0].set_title("Loss (smoothed)")
    axs[0].set_xlabel("Step"); axs[0].set_ylabel("Loss")
    axs[0].legend(); axs[0].grid(True)
    axs[1].plot(df['step'], df['lr_G'], label='LR Generator')
    axs[1].plot(df['step'], df['lr_D'], label='LR Discriminator')
    axs[1].set_title("Learning Rates")
    axs[1].set_xlabel("Step"); axs[1].set_ylabel("LR")
    axs[1].legend(); axs[1].grid(True)
    axs[2].plot(df['step'], smooth(df['grad_norm_G']), label='GradNorm G')
    axs[2].plot(df['step'], smooth(df['grad_norm_D']), label='GradNorm D')
    axs[2].set_title("Gradient Norms (smoothed)")
    axs[2].set_xlabel("Step"); axs[2].set_ylabel("Norm")
    axs[2].legend(); axs[2].grid(True)
    plt.suptitle("GAN Training Diagnostics", fontsize=16)
    plt.show()

# log_step (TensorBoard + 간단 출력)
def log_step(step, loss_components, optimizer_G, optimizer_D, G, D, inputs, targets, fake):
    # loss
    writer.add_scalar("loss/D_LSGAN", loss_components["loss_D_LSGAN"].item(), step)
    writer.add_scalar("loss/G_LSGAN", loss_components["loss_G_LSGAN"].item(), step)
    writer.add_scalar("loss/FeatureMatching", loss_components["loss_FM"].item(), step)
    writer.add_scalar("loss/Correlation", loss_components["loss_CC"].item(), step)
    writer.add_scalar("loss/G_total", loss_components["loss_G_total"].item(), step)
    # lr
    writer.add_scalar("lr/Generator", optimizer_G.param_groups[0]["lr"], step)
    writer.add_scalar("lr/Discriminator", optimizer_D.param_groups[0]["lr"], step)
    # grad norms
    grad_norm_G = sum(p.grad.norm().item()**2 for p in G.parameters() if p.grad is not None) ** 0.5
    grad_norm_D = sum(p.grad.norm().item()**2 for p in D.parameters() if p.grad is not None) ** 0.5
    writer.add_scalar("grad_norm/Generator", grad_norm_G, step)
    writer.add_scalar("grad_norm/Discriminator", grad_norm_D, step)
    # images (middle slice)
    depth_idx = inputs.shape[2] // 2
    def norm_img(tensor):
        img = tensor[0, 0, depth_idx].detach().cpu()
        img = (img - img.min()) / (img.max() - img.min() + 1e-6)
        return img.unsqueeze(0)
    writer.add_image("slice/Input", norm_img(inputs), step)
    writer.add_image("slice/Target", norm_img(targets), step)
    writer.add_image("slice/Fake", norm_img(fake), step)
    writer.add_image("slice/Residual", norm_img(fake - targets), step)
```

### 4. Overfit 테스트: 아주 작은 서브셋(한두 배치)으로 손실이 줄어드는지 확인

Generator/Discriminator가 제대로 신호를 받고 역전파가 작동하는지 검증한다.  
Loss breakdown도 함께 출력.


```python
import torch
import torch.nn as nn
import torch.nn.functional as F

def compute_gan_loss_components(D, G, input_tensor, target_tensor, opt):
    """
    Returns breakdown of losses for 3D Pix2PixCC-style GAN.
    """
    device = next(G.parameters()).device
    dtype = torch.float16 if getattr(opt, "data_type", 32) == 16 else torch.float32

    MSE = nn.MSELoss()
    FMcriterion = nn.L1Loss()
    n_D = opt.n_D

    fake = G(input_tensor)

    # -------- real / fake pair with channel balance (for discriminator) --------
    if opt.ch_balance > 0:
        # construct balanced real/fake pairs (detached for fake)
        real_pair, fake_pair_detached = _balanced_pairs_for_debug(input_tensor, target_tensor, fake, opt, detach_fake=True)
        real_features = D(real_pair)
        fake_features_detached = D(fake_pair_detached)
    else:
        real_features = D(torch.cat((input_tensor, target_tensor), dim=1))
        fake_features_detached = D(torch.cat((input_tensor, fake.detach()), dim=1))

    # Discriminator LSGAN loss
    loss_D_LSGAN = 0.0
    for i in range(n_D):
        real_grid = get_grid(real_features[i][-1], is_real=True).to(device, dtype=dtype)
        fake_grid = get_grid(fake_features_detached[i][-1], is_real=False).to(device, dtype=dtype)
        loss_D_LSGAN += (MSE(real_features[i][-1], real_grid) + MSE(fake_features_detached[i][-1], fake_grid)) * 0.5

    # -------- generator side: prepare features (with or without channel balance) --------
    if opt.ch_balance > 0:
        fake_pair = _make_fake_pair_for_debug(input_tensor, fake, opt)
        real_pair = _make_real_pair_for_debug(input_tensor, target_tensor, opt)
        fake_features = D(fake_pair)
        real_features_for_FM = D(real_pair)
    else:
        fake_features = D(torch.cat((input_tensor, fake), dim=1))
        real_features_for_FM = D(torch.cat((input_tensor, target_tensor), dim=1))

    # Generator LSGAN and Feature Matching
    loss_G_LSGAN = 0.0
    loss_FM = 0.0
    for i in range(n_D):
        real_grid_for_G = get_grid(fake_features[i][-1], is_real=True).to(device, dtype=dtype)
        loss_G_LSGAN += MSE(fake_features[i][-1], real_grid_for_G) * 0.5 * opt.lambda_LSGAN
        for j in range(len(fake_features[i])):
            loss_FM += FMcriterion(fake_features[i][j], real_features_for_FM[i][j].detach())
    loss_FM = loss_FM * (1.0 / n_D) * opt.lambda_FM

    # Correlation / CCC loss
    loss_CC = 0.0
    for i in range(opt.n_CC):
        real_down = target_tensor.to(device, dtype=dtype)
        fake_down = fake.to(device, dtype=dtype)
        for _ in range(i):
            real_down = F.avg_pool3d(real_down, kernel_size=3, stride=2, padding=1, count_include_pad=False)
            fake_down = F.avg_pool3d(fake_down, kernel_size=3, stride=2, padding=1, count_include_pad=False)
        # inspector logic
        rd = real_down - torch.mean(real_down)
        fd = fake_down - torch.mean(fake_down)
        r_num = torch.sum(rd * fd)
        r_den = torch.sqrt(torch.sum(rd ** 2)) * torch.sqrt(torch.sum(fd ** 2))
        PCC_val = r_num / (r_den + opt.eps)
        if getattr(opt, "ccc", False):
            numerator = 2 * PCC_val * torch.std(real_down) * torch.std(fake_down)
            denominator = (torch.var(real_down) + torch.var(fake_down) +
                           (torch.mean(real_down) - torch.mean(fake_down)) ** 2)
            CCC_val = numerator / (denominator + opt.eps)
            loss_CC_i = 1.0 - CCC_val
        else:
            loss_CC_i = 1.0 - PCC_val
        loss_CC += loss_CC_i * (1.0 / opt.n_CC) * opt.lambda_CC

    # Total generator loss
    loss_G_total = loss_G_LSGAN + loss_FM + loss_CC

    return {
        "loss_D_LSGAN": loss_D_LSGAN,
        "loss_G_LSGAN": loss_G_LSGAN,
        "loss_FM": loss_FM,
        "loss_CC": loss_CC,
        "loss_G_total": loss_G_total,
        "fake": fake
    }


# Helper functions (reuse logic from your implementation)
def _balanced_pairs_for_debug(inp, target, fake, opt, detach_fake=False):
    if opt.ch_balance <= 0:
        real_pair = torch.cat((inp, target), dim=1)
        fake_pair = torch.cat((inp, fake.detach() if detach_fake else fake), dim=1)
        return real_pair, fake_pair
    ch_ratio = float(opt.input_ch) / float(opt.target_ch) * opt.ch_balance
    real_pair = torch.cat((inp, target), dim=1)
    fake_pair = torch.cat((inp, fake.detach() if detach_fake else fake), dim=1)
    if ch_ratio > 1:
        for _ in range(int(ch_ratio) - 1):
            real_pair = torch.cat((real_pair, target), dim=1)
            fake_pair = torch.cat((fake_pair, fake.detach() if detach_fake else fake), dim=1)
    elif ch_ratio < 1:
        for _ in range(int(1 / ch_ratio) - 1):
            real_pair = torch.cat((inp, real_pair), dim=1)
            fake_pair = torch.cat((inp, fake_pair), dim=1)
    return real_pair, fake_pair

def _make_fake_pair_for_debug(inp, fake, opt):
    if opt.ch_balance <= 0:
        return torch.cat((inp, fake), dim=1)
    ch_ratio = float(opt.input_ch) / float(opt.target_ch) * opt.ch_balance
    fake_pair = torch.cat((inp, fake), dim=1)
    if ch_ratio > 1:
        for _ in range(int(ch_ratio) - 1):
            fake_pair = torch.cat((fake_pair, fake), dim=1)
    elif ch_ratio < 1:
        for _ in range(int(1 / ch_ratio) - 1):
            fake_pair = torch.cat((inp, fake_pair), dim=1)
    return fake_pair

def _make_real_pair_for_debug(inp, target, opt):
    if opt.ch_balance <= 0:
        return torch.cat((inp, target), dim=1)
    ch_ratio = float(opt.input_ch) / float(opt.target_ch) * opt.ch_balance
    real_pair = torch.cat((inp, target), dim=1)
    if ch_ratio > 1:
        for _ in range(int(ch_ratio) - 1):
            real_pair = torch.cat((real_pair, target), dim=1)
    elif ch_ratio < 1:
        for _ in range(int(1 / ch_ratio) - 1):
            real_pair = torch.cat((inp, real_pair), dim=1)
    return real_pair
```

```python
# 모델 초기화 (opt가 정의돼 있다고 가정)
from models.gan.model import GeneratorPix2PixCC3D, MultiScaleDiscriminator3D, Pix2PixCCLoss3D
from models.gan.utils import weights_init, get_grid  # 필요한 유틸 임포트

# build opt-like
def build_opt():
    class O: pass
    o = O()
    o.input_ch = 1; o.target_ch = 1
    o.n_gf = 32; o.n_df = 32
    o.n_downsample = 2; o.n_residual = 4
    o.trans_conv = False
    o.padding_type = "replication"
    o.norm_type = "InstanceNorm3d"
    o.ch_balance = 0.0
    o.n_D = 2
    o.lambda_LSGAN = 1.0
    o.lambda_FM = 10.0
    o.lambda_CC = 1.0
    o.n_CC = 1
    o.ccc = False
    o.eps = 1e-8
    o.gpu_ids = 0
    o.data_type = 32
    return o

opt = build_opt()
opt.model_dir = "debug_models"
opt.image_dir = "debug_images"
os.makedirs(opt.model_dir, exist_ok=True)
os.makedirs(opt.image_dir, exist_ok=True)

G = GeneratorPix2PixCC3D(opt).to(device)
D = MultiScaleDiscriminator3D(opt).to(device)
G.apply(weights_init); D.apply(weights_init)
criterion = Pix2PixCCLoss3D(opt)

# Optimizers
optimizer_G = Adam(G.parameters(), lr=1e-4, betas=(0.5,0.999))
optimizer_D = Adam(D.parameters(), lr=1e-4, betas=(0.5,0.999))
scheduler_G = CosineAnnealingLR(optimizer_G, T_max=10)
scheduler_D = CosineAnnealingLR(optimizer_D, T_max=10)

# 한 배치 overfit: 5 step
raw_x, raw_y = next(iter(loader))
x, y = prepare_batch(raw_x, raw_y, G)

print("Starting overfit test (5 steps)...")
for step in range(5):
    components = compute_gan_loss_components(D, G, x, y, opt)
    loss_G = components["loss_G_total"]
    loss_D = components["loss_D_LSGAN"]

    optimizer_G.zero_grad()
    loss_G.backward()
    torch.nn.utils.clip_grad_norm_(G.parameters(), 5.0)
    optimizer_G.step()

    optimizer_D.zero_grad()
    loss_D.backward()
    torch.nn.utils.clip_grad_norm_(D.parameters(), 5.0)
    optimizer_D.step()

    print(f"[Overfit Step {step}] G_loss={loss_G.item():.4f}, D_loss={loss_D.item():.4f}")
    log_step(step, components, optimizer_G, optimizer_D, G, D, x, y, components["fake"])
```

### 5. Batch size sweep

여러 배치 크기로 forward + 한 스텝 수행하면서 시간, 메모리(allocated/peak) 등을 측정하여 실전 학습에 적절한 batch size를 선택한다.


```python
def safe_profile_batch_size(sizes, loader, G, D, opt):
    results = []
    for bs in sizes:
        try:
            tmp_loader = DataLoader(loader.dataset, batch_size=bs, shuffle=False, num_workers=0)
            raw_x, raw_y = next(iter(tmp_loader))
            x, y = prepare_batch(raw_x, raw_y, G)

            torch.cuda.empty_cache()
            torch.cuda.reset_peak_memory_stats()

            start = time.time()
            components = compute_gan_loss_components(D, G, x, y, opt)
            loss_G = components["loss_G_total"]
            loss_D = components["loss_D_LSGAN"]
            end = time.time()

            # lightweight backward to exercise grads
            optimizer_G.zero_grad(); loss_G.backward(retain_graph=True); optimizer_G.step()
            optimizer_D.zero_grad(); loss_D.backward(); optimizer_D.step()

            alloc = torch.cuda.memory_allocated()
            reserved = torch.cuda.memory_reserved()
            peak_alloc = torch.cuda.max_memory_allocated()
            peak_reserved = torch.cuda.max_memory_reserved()
            results.append({
                "batch_size": bs,
                "time_s": end - start,
                "loss_G": loss_G.item(),
                "loss_D": loss_D.item(),
                "alloc_GB": alloc / 1e9,
                "reserved_GB": reserved / 1e9,
                "peak_alloc_GB": peak_alloc / 1e9,
                "peak_reserved_GB": peak_reserved / 1e9,
            })
            print(f"✅ bs={bs} | time {end-start:.3f}s | G_loss {loss_G.item():.3f} D_loss {loss_D.item():.3f} | peak_alloc {peak_alloc/1e9:.3f} GB")
        except RuntimeError as e:
            if "out of memory" in str(e):
                print(f"❌ bs={bs} OOM, skipping.")
                torch.cuda.empty_cache()
            else:
                raise
    return pd.DataFrame(results)

import torch
torch.cuda.empty_cache()

batch_sizes = [16, 8, 4, 2, 1]
df_profile = safe_profile_batch_size(batch_sizes, loader, G, D, opt)
df_profile
```

```python
import torch
torch.cuda.empty_cache()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 모델이 현재 디바이스에 올라가 있는지 확인
G = G.to(device)
D = D.to(device)

# 한 배치만 가져오기
x, y = next(iter(loader))
x = x.float().to(device)
y = y.float().to(device)

with torch.no_grad():
    try:
        fake = G(x)  # forward only
        print("✅ Generator forward success:", fake.shape)
    except RuntimeError as e:
        print("❌ Forward failed:", str(e).splitlines()[0])
```

```python
from torch.cuda.amp import autocast, GradScaler
import torch.nn as nn

# 준비
G.train(); D.train()
scaler = GradScaler()
optimizer_G = torch.optim.Adam(G.parameters(), lr=1e-4, betas=(0.5, 0.999))
optimizer_D = torch.optim.Adam(D.parameters(), lr=1e-4, betas=(0.5, 0.999))

# 한 배치 가져오기
x, y = next(iter(loader))
x = x.float().to(device)
y = y.float().to(device)

# 메모리 초기화 (선택)
torch.cuda.empty_cache()
torch.cuda.reset_peak_memory_stats()

# forward + loss breakdown with AMP
with autocast():
    comps = compute_gan_loss_components(D, G, x, y, opt)
    loss_G = comps["loss_G_total"]
    loss_D = comps["loss_D_LSGAN"]
    loss_FM = comps["loss_FM"]
    loss_CC = comps["loss_CC"]
    loss_G_LSGAN = comps["loss_G_LSGAN"]

# generator update
optimizer_G.zero_grad()
scaler.scale(loss_G).backward(retain_graph=True)  # retain if you inspect D after
scaler.step(optimizer_G)
scaler.update()

# discriminator update
optimizer_D.zero_grad()
scaler.scale(loss_D).backward()
scaler.step(optimizer_D)
scaler.update()

# gradient norms (optional)
def grad_norm(model):
    return torch.sqrt(sum(p.grad.detach().norm()**2 for p in model.parameters() if p.grad is not None))

print(f"=== Overfit step ===")
print(f"Loss D LSGAN: {loss_D.item():.4f}")
print(f"Loss G total: {loss_G.item():.4f} (LSGAN {loss_G_LSGAN.item():.4f}, FM {loss_FM.item():.4f}, CC {loss_CC.item():.4f})")
print(f"Grad norm G: {grad_norm(G):.4f}, D: {grad_norm(D):.4f}")
print(f"Peak alloc (GB): {torch.cuda.max_memory_allocated()/1e9:.3f}")
```

```python
num_steps = 10
for step in range(num_steps):
    with autocast(device_type="cuda"):
        comps = compute_gan_loss_components(D, G, x, y, opt)
        loss_G = comps["loss_G_total"]
        loss_D = comps["loss_D_LSGAN"]

    optimizer_G.zero_grad()
    scaler.scale(loss_G).backward(retain_graph=True)
    scaler.unscale_(optimizer_G)
    torch.nn.utils.clip_grad_norm_(G.parameters(), 1.0)
    scaler.step(optimizer_G)
    scaler.update()

    optimizer_D.zero_grad()
    scaler.scale(loss_D).backward()
    scaler.unscale_(optimizer_D)
    torch.nn.utils.clip_grad_norm_(D.parameters(), 1.0)
    scaler.step(optimizer_D)
    scaler.update()

    if step % 2 == 0:
        print(f"[Step {step}] G_loss={loss_G.item():.4f} D_loss={loss_D.item():.4f}")
```

### 6. 짧은 multi-step 학습 및 진단 기록

초기 hyperparameter/균형 상태를 보기 위해 몇십 step만 돌려서 손실, learning rate, gradient norm을 기록한다.


```python
records = []
global_step = 0
max_steps = 50  # 테스트용

G.train(); D.train()
for step in range(max_steps):
    try:
        raw_x, raw_y = next(iter(loader))
    except StopIteration:
        break
    x, y = prepare_batch(raw_x, raw_y, G)
    components = compute_gan_loss_components(D, G, x, y, opt)
    loss_G = components["loss_G_total"]
    loss_D = components["loss_D_LSGAN"]

    optimizer_G.zero_grad()
    loss_G.backward()
    torch.nn.utils.clip_grad_norm_(G.parameters(), 5.0)
    optimizer_G.step()

    optimizer_D.zero_grad()
    loss_D.backward()
    torch.nn.utils.clip_grad_norm_(D.parameters(), 5.0)
    optimizer_D.step()

    # scheduler (per epoch or per step depending on 설정)
    scheduler_G.step()
    scheduler_D.step()

    # gradient norms
    grad_norm_G = sum(p.grad.norm().item()**2 for p in G.parameters() if p.grad is not None) ** 0.5
    grad_norm_D = sum(p.grad.norm().item()**2 for p in D.parameters() if p.grad is not None) ** 0.5

    lr_G = optimizer_G.param_groups[0]["lr"]
    lr_D = optimizer_D.param_groups[0]["lr"]

    records.append({
        "step": global_step,
        "loss_G": loss_G.item(),
        "loss_D": loss_D.item(),
        "lr_G": lr_G,
        "lr_D": lr_D,
        "grad_norm_G": grad_norm_G,
        "grad_norm_D": grad_norm_D,
    })

    if global_step % 10 == 0:
        print(f"[Step {global_step}] G_loss={loss_G.item():.4f}, D_loss={loss_D.item():.4f}, lr_G={lr_G:.2e}")

    log_step(global_step, components, optimizer_G, optimizer_D, G, D, x, y, components["fake"])
    global_step += 1

# DataFrame으로 보고
df = pd.DataFrame(records)
plot_gan_diagnostics(df)
```

### 7. 슬라이스 기반 예측 품질 시각화

중간 depth에 대해 입력, 타깃, 생성물, residual을 로그 변환 포함해 비교하고 MAE/RMSE도 계산한다.


```python
def signed_log10(x, eps=1e-6):
    return torch.sign(x) * torch.log10(1 + torch.abs(x) + eps)

# 배치 0, 채널 0, 중간 슬라이스
depth_idx = x.shape[2] // 2
inp = x[0,0,depth_idx].detach().cpu()
tgt = y[0,0,depth_idx].detach().cpu()
fake = components["fake"][0,0,depth_idx].detach().cpu()

inp_log = signed_log10(inp).numpy()
tgt_log = signed_log10(tgt).numpy()
fake_log = signed_log10(fake).numpy()

# 에러
error = fake - tgt
mae = torch.mean(torch.abs(error)).item()
rmse = torch.sqrt(torch.mean(error**2)).item()

fig, axs = plt.subplots(2, 3, figsize=(15, 8))
axs[0,0].imshow(inp_log, cmap='viridis'); axs[0,0].set_title('Input signed log')
axs[0,1].imshow(tgt_log, cmap='viridis'); axs[0,1].set_title('Target signed log')
axs[0,2].imshow(fake_log, cmap='viridis'); axs[0,2].set_title('Fake signed log')
axs[1,0].imshow(torch.abs(error).numpy(), cmap='magma'); axs[1,0].set_title('Abs Error')
axs[1,1].imshow((error**2).numpy(), cmap='magma'); axs[1,1].set_title('Squared Error')
res_log = signed_log10(error).numpy()
im = axs[1,2].imshow(res_log, cmap='bwr', vmin=-np.max(np.abs(res_log)), vmax=np.max(np.abs(res_log)))
axs[1,2].set_title('Residual signed log')
for ax in axs.flatten():
    ax.axis('off')
fig.suptitle(f"Slice diagnostics (z={depth_idx}) | MAE={mae:.4e}, RMSE={rmse:.4e}", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn.functional as F

def compute_metrics(target, pred):
    """
    target, pred: numpy arrays of the same shape (e.g., 2D slice)
    Returns MAE and RMSE.
    """
    diff = pred - target
    mae = np.mean(np.abs(diff))
    rmse = np.sqrt(np.mean(diff**2))
    return mae, rmse

def plot_slice_diagnostics(
    input_vol, target_vol, fake_vol,
    slice_idx=30,
    config=None,  # dict of experiment metadata
    log_transform=True,
    signed=True,
    cmap='viridis',
):
    """
    input_vol, target_vol, fake_vol: torch tensors of shape (B, C, D, H, W) or (C, D, H, W)
    config: dictionary with experiment settings to annotate
    """
    # take first batch / channel
    def to_numpy(t):
        if isinstance(t, torch.Tensor):
            t = t.detach().cpu()
        arr = t.numpy()
        if arr.ndim == 5:  # (B,C,D,H,W)
            arr = arr[0, 0]
        elif arr.ndim == 4:  # (C,D,H,W)
            arr = arr[0]
        return arr  # (D,H,W)

    inp = to_numpy(input_vol)
    tgt = to_numpy(target_vol)
    fake = to_numpy(fake_vol)

    slice_inp = inp[slice_idx]
    slice_tgt = tgt[slice_idx]
    slice_fake = fake[slice_idx]

    # optional signed log: log10(1 + sign(x)*|x|)
    def signed_log(x):
        return np.sign(x) * np.log10(1 + np.abs(x))

    if log_transform:
        slice_inp_plot = signed_log(slice_inp) if signed else np.log10(1 + slice_inp)
        slice_tgt_plot = signed_log(slice_tgt) if signed else np.log10(1 + slice_tgt)
        slice_fake_plot = signed_log(slice_fake) if signed else np.log10(1 + slice_fake)
    else:
        slice_inp_plot = slice_inp
        slice_tgt_plot = slice_tgt
        slice_fake_plot = slice_fake

    abs_err = np.abs(slice_tgt - slice_fake)
    sq_err = (slice_tgt - slice_fake) ** 2
    residual = slice_fake_plot - slice_tgt_plot

    mae, rmse = compute_metrics(slice_tgt, slice_fake)

    # Plotting
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    fig.suptitle(
        f"Slice diagnostics (z={slice_idx}) | MAE={mae:.4e}, RMSE={rmse:.4e}",
        fontsize=16,
    )

    titles = [
        "Input signed log" if log_transform and signed else "Input",
        "Target signed log" if log_transform and signed else "Target",
        "Fake signed log" if log_transform and signed else "Fake",
        "Abs Error",
        "Squared Error",
        "Residual signed log" if log_transform and signed else "Residual",
    ]
    images = [
        slice_inp_plot,
        slice_tgt_plot,
        slice_fake_plot,
        abs_err,
        sq_err,
        residual,
    ]

    for ax, im, title in zip(axes.flatten(), images, titles):
        im_obj = ax.imshow(im, cmap=cmap)
        ax.set_title(title, fontsize=12)
        ax.axis("off")
        fig.colorbar(im_obj, ax=ax, fraction=0.046, pad=0.04)

    # Annotate config as text box below
    if config is None:
        config = {}
    # suptitle 이후, 전체 하단에 붙이기
    config_str = " | ".join(f"{k}={v}" for k, v in config.items())
    fig.text(
        0.5,
        0.005,
        config_str,
        ha="center",
        va="bottom",
        fontsize=20,
        wrap=True,
    )


    plt.tight_layout(rect=[0, 0.05, 1, 0.95])
    plt.show()
```

```python
from torch.amp import autocast, GradScaler
import time
import pandas as pd
import torch.nn as nn

def grad_norm(model):
    return torch.sqrt(sum(p.grad.detach().norm()**2 for p in model.parameters() if p.grad is not None))

scaler = GradScaler(device="cuda")
optimizer_G = torch.optim.Adam(G.parameters(), lr=1e-5, betas=(0.5, 0.999))
optimizer_D = torch.optim.Adam(D.parameters(), lr=1e-5, betas=(0.5, 0.999))

# dataset/loader metadata
batch_size = loader.batch_size if hasattr(loader, "batch_size") else 2
dataset_size = len(loader.dataset)
steps_per_epoch = (dataset_size + batch_size - 1) // batch_size  # ceil

records = []
num_epochs = 1  # 짧은 테스트
max_global_steps = 50
global_step = 0

for epoch in range(1, num_epochs + 1):
    for batch_idx, (x, y) in enumerate(loader):
        if global_step >= max_global_steps:
            break

        x = x.float().to(device)
        y = y.float().to(device)

        with autocast(device_type="cuda"):
            comps = compute_gan_loss_components(D, G, x, y, opt)
            loss_G = comps["loss_G_total"]
            loss_D = comps["loss_D_LSGAN"]
            loss_FM = comps.get("loss_FM", torch.tensor(0.0, device=loss_G.device))
            loss_CC = comps.get("loss_CC", torch.tensor(0.0, device=loss_G.device))
            loss_G_LSGAN = comps.get("loss_G_LSGAN", torch.tensor(0.0, device=loss_G.device))

        # Generator update
        optimizer_G.zero_grad()
        scaler.scale(loss_G).backward(retain_graph=True)
        scaler.unscale_(optimizer_G)
        torch.nn.utils.clip_grad_norm_(G.parameters(), 1.0)
        scaler.step(optimizer_G)
        scaler.update()

        # Discriminator update
        optimizer_D.zero_grad()
        scaler.scale(loss_D).backward()
        scaler.unscale_(optimizer_D)
        torch.nn.utils.clip_grad_norm_(D.parameters(), 1.0)
        scaler.step(optimizer_D)
        scaler.update()

        # Metrics
        duration = time.time() - start_time if (start_time := time.time()) else 0.0
        gn_G = grad_norm(G).item()
        gn_D = grad_norm(D).item()
        current_lr = optimizer_G.param_groups[0]["lr"]
        peak_alloc = torch.cuda.max_memory_allocated() / 1e9
        peak_reserved = torch.cuda.max_memory_reserved() / 1e9

        records.append({
            "epoch": epoch,
            "batch_idx": batch_idx,
            "global_step": global_step,
            "G_loss": loss_G.item(),
            "D_loss": loss_D.item(),
            "LSGAN_G": loss_G_LSGAN.item(),
            "FM": loss_FM.item(),
            "CC": loss_CC.item(),
            "grad_norm_G": gn_G,
            "grad_norm_D": gn_D,
            "lr": current_lr,
            "time_s": duration,
            "peak_alloc_GB": peak_alloc,
            "peak_reserved_GB": peak_reserved,
        })

        if global_step % 5 == 0:
            print(f"[Epoch {epoch} Batch {batch_idx} | global_step={global_step}] G_loss={loss_G.item():.4f} D_loss={loss_D.item():.4f}")

        global_step += 1
    if global_step >= max_global_steps:
        break

df = pd.DataFrame(records)
df.to_csv("overfit_diagnostics.csv", index=False)
```
