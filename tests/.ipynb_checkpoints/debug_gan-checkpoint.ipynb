```python
import time
import pandas as pd
import torch

def test_batch_sizes_gan_logging(batch_sizes):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    records = []

    for bs in batch_sizes:
        try:
            loader = DataLoader(dataset, batch_size=bs, shuffle=False, num_workers=0)

            # fresh models per trial
            G_test = GeneratorPix2PixCC3D(opt).to(device)
            D_test = MultiScaleDiscriminator3D(opt).to(device)
            G_test.apply(weights_init)
            D_test.apply(weights_init)
            criterion_test = Pix2PixCCLoss3D(opt)

            x, y = next(iter(loader))
            if x.ndim == 4:
                x = x[:, None]
            if y.ndim == 4:
                y = y[:, None]
            x = x.float().to(device)
            y = y.float().to(device)

            torch.cuda.reset_peak_memory_stats(device)
            start = time.time()
            with torch.no_grad():
                loss_D, loss_G, _, fake = criterion_test(D_test, G_test, x, y)
            duration = time.time() - start

            allocated = torch.cuda.memory_allocated(device)
            reserved = torch.cuda.memory_reserved(device)
            peak_alloc = torch.cuda.max_memory_allocated(device)
            peak_reserved = torch.cuda.max_memory_reserved(device)

            records.append({
                "batch_size": bs,
                "loss_D": loss_D.item(),
                "loss_G": loss_G.item(),
                "time_s": duration,
                "allocated_GB": allocated / 1e9,
                "reserved_GB": reserved / 1e9,
                "peak_alloc_GB": peak_alloc / 1e9,
                "peak_reserved_GB": peak_reserved / 1e9,
            })

            print(f"✅ bs={bs} | loss_D={loss_D.item():.4f} loss_G={loss_G.item():.4f} time={duration:.3f}s "
                  f"peak_alloc={peak_alloc/1e9:.3f}GB")
        except RuntimeError as e:
            msg = str(e).splitlines()[0]
            print(f"❌ bs={bs} failed: {msg}")
            torch.cuda.empty_cache()
            records.append({
                "batch_size": bs,
                "loss_D": None,
                "loss_G": None,
                "time_s": None,
                "allocated_GB": None,
                "reserved_GB": None,
                "peak_alloc_GB": None,
                "peak_reserved_GB": None,
                "error": msg,
            })

    df = pd.DataFrame(records)
    display(df.sort_values("batch_size", ascending=False))
    return df
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# 가정: x, y, opt, G, D가 이미 정의되어 있고 G/D는 eval/train 상태 적절히
# 그리고 get_grid, Pix2PixCCLoss3D 등이 import 되어 있어야 함:
from models.gan.model import Pix2PixCCLoss3D
from models.gan.model import GeneratorPix2PixCC3D, MultiScaleDiscriminator3D
from models.gan.utils import get_grid

def compute_gan_loss_components(D, G, x, y, opt):
    device = next(G.parameters()).device
    dtype = torch.float16 if getattr(opt, "data_type", 32) == 16 else torch.float32

    mse = nn.MSELoss()
    fm_crit = nn.L1Loss()
    loss_wrapper = Pix2PixCCLoss3D(opt)

    fake = G(x)

    # ---- Discriminator real/fake features ----
    if opt.ch_balance > 0:
        real_pair, fake_pair = loss_wrapper._balanced_pairs(x, y, fake)
        real_feats = D(real_pair)
        fake_feats_detached = D(fake_pair.detach())
    else:
        real_feats = D(torch.cat((x, y), dim=1))
        fake_feats_detached = D(torch.cat((x, fake.detach()), dim=1))

    # Discriminator LSGAN loss
    loss_D_LSGAN = 0.0
    for i in range(opt.n_D):
        real_grid = get_grid(real_feats[i][-1], is_real=True).to(device, dtype)
        fake_grid = get_grid(fake_feats_detached[i][-1], is_real=False).to(device, dtype)
        loss_D_LSGAN += (mse(real_feats[i][-1], real_grid) + mse(fake_feats_detached[i][-1], fake_grid)) * 0.5

    # ---- Generator side features ----
    if opt.ch_balance > 0:
        fake_pair_full = loss_wrapper._make_fake_pair(x, fake)
        real_pair_full = loss_wrapper._make_real_pair(x, y)
        fake_feats = D(fake_pair_full)
        real_feats_for_FM = D(real_pair_full)
    else:
        fake_feats = D(torch.cat((x, fake), dim=1))
        real_feats_for_FM = D(torch.cat((x, y), dim=1))

    # Generator LSGAN
    loss_G_LSGAN = 0.0
    for i in range(opt.n_D):
        real_grid = get_grid(fake_feats[i][-1], is_real=True).to(device, dtype)
        loss_G_LSGAN += mse(fake_feats[i][-1], real_grid) * 0.5 * opt.lambda_LSGAN

    # Feature Matching
    loss_FM = 0.0
    for i in range(opt.n_D):
        for j in range(len(fake_feats[i])):
            loss_FM += fm_crit(fake_feats[i][j], real_feats_for_FM[i][j].detach())
    loss_FM = loss_FM * (1.0 / opt.n_D) * opt.lambda_FM

    # Correlation / CCC
    loss_CC = 0.0
    for i in range(opt.n_CC):
        real_down = y.to(device, dtype)
        fake_down = fake.to(device, dtype)
        for _ in range(i):
            real_down = F.avg_pool3d(real_down, kernel_size=3, stride=2, padding=1)
            fake_down = F.avg_pool3d(fake_down, kernel_size=3, stride=2, padding=1)
        loss_CC_part = loss_wrapper._inspector(real_down, fake_down)
        loss_CC += loss_CC_part * (1.0 / opt.n_CC) * opt.lambda_CC

    loss_G_total = loss_G_LSGAN + loss_FM + loss_CC

    return {
        "loss_D_LSGAN": loss_D_LSGAN,
        "loss_G_LSGAN": loss_G_LSGAN,
        "loss_FM": loss_FM,
        "loss_CC": loss_CC,
        "loss_G_total": loss_G_total,
        "loss_D_total": loss_D_LSGAN,
        "fake": fake,
    }


# 실행 예
components = compute_gan_loss_components(D, G, x, y, opt)
print("=== GAN loss components ===")
print(f"Discriminator LSGAN: {components['loss_D_LSGAN'].item():.4f}")
print(f"Generator LSGAN:     {components['loss_G_LSGAN'].item():.4f}")
print(f"Feature Matching:    {components['loss_FM'].item():.4f}")
print(f"Correlation/CCC:     {components['loss_CC'].item():.4f}")
print(f"Total Generator:     {components['loss_G_total'].item():.4f}")
```

```python
# GAN용: Generator / Discriminator 각각 optimizer, scheduler
from torch.optim import Adam
from torch.optim.lr_scheduler import CosineAnnealingLR

# learning rates는 필요하면 따로 설정 가능
lr_G = 1e-4
lr_D = 1e-4

optimizer_G = Adam(G.parameters(), lr=lr_G, betas=(0.5, 0.999))
optimizer_D = Adam(D.parameters(), lr=lr_D, betas=(0.5, 0.999))

# Cosine annealing schedule (예: 10 epoch 주기)
scheduler_G = CosineAnnealingLR(optimizer_G, T_max=10)
scheduler_D = CosineAnnealingLR(optimizer_D, T_max=10)

print("✅ Optimizers and cosine annealing schedulers initialized for G and D.")
```

```python
# Cell: 빠른 GAN 학습 루프 (1 epoch당 일부 batch만, G/D 업데이트)
from tqdm import tqdm

# 전제: G, D, opt, dataset/loader, criterion = Pix2PixCCLoss3D(opt),
# optimizer_G, optimizer_D, scheduler_G, scheduler_D가 이미 정의되어 있고 device도 설정됨.

G.train()
D.train()
n_batch = 10  # 테스트용으로 10개 배치만

for epoch in range(3):
    epoch_loss_G = 0.0
    epoch_loss_D = 0.0

    loop = tqdm(loader, desc=f"Epoch {epoch+1}", total=n_batch)
    for i, (inputs, targets) in enumerate(loop):
        if i >= n_batch:
            break

        # 채널 차원 보정: (B, D, H, W) -> (B,1,D,H,W)
        if inputs.ndim == 4:
            inputs = inputs[:, None]
        if targets.ndim == 4:
            targets = targets[:, None]

        inputs = inputs.float().to(device)
        targets = targets.float().to(device)

        # --- Forward & loss ---
        loss_D, loss_G, _, fake = criterion(D, G, inputs, targets)

        # --- Update Generator ---
        optimizer_G.zero_grad()
        loss_G.backward(retain_graph=True)  # D loss will use detached; retain if needed
        optimizer_G.step()

        # --- Update Discriminator ---
        optimizer_D.zero_grad()
        loss_D.backward()
        optimizer_D.step()

        epoch_loss_G += loss_G.item()
        epoch_loss_D += loss_D.item()

        loop.set_postfix({
            "G_loss": f"{loss_G.item():.4f}",
            "D_loss": f"{loss_D.item():.4f}"
        })

    # Scheduler step (per epoch)
    scheduler_G.step()
    scheduler_D.step()

    avg_G = epoch_loss_G / n_batch
    avg_D = epoch_loss_D / n_batch
    current_lr_G = scheduler_G.get_last_lr()[0]
    current_lr_D = scheduler_D.get_last_lr()[0]

    print(f"📉 Epoch {epoch+1} (partial) | Avg G_loss: {avg_G:.4f} | Avg D_loss: {avg_D:.4f} "
          f"| LR_G: {current_lr_G:.2e} | LR_D: {current_lr_D:.2e}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torch

def signed_log10(x, eps=1e-6):
    """음수도 표현 가능한 log 변환: sign(x) * log10(1 + |x|)"""
    return torch.sign(x) * torch.log10(1 + torch.abs(x) + eps)

# assume inputs, targets, fake are tensors from GAN loop, shape [B, C, D, H, W]
# pick batch 0, channel 0, slice at depth index (e.g., middle)
depth_idx = inputs.shape[2] // 2  # z=30 if D=60
inp = inputs[0, 0, depth_idx].detach().cpu()
tgt = targets[0, 0, depth_idx].detach().cpu()
pred = fake[0, 0, depth_idx].detach().cpu()

# apply log-transform in a signed way to preserve sign information
inp_log = signed_log10(inp).numpy()
tgt_log = signed_log10(tgt).numpy()
pred_log = signed_log10(pred).numpy()

# compute simple error maps (on original scale)
error = (pred - tgt)
abs_error = torch.abs(error).numpy()
squared_error = (error ** 2).numpy()

# aggregate metrics
mae = np.mean(np.abs(pred.numpy() - tgt.numpy()))
rmse = np.sqrt(np.mean((pred.numpy() - tgt.numpy())**2))

# plotting
fig, axs = plt.subplots(2, 3, figsize=(15, 8))

# first row: log-scaled fields
axs[0, 0].imshow(inp_log, cmap='viridis')
axs[0, 0].set_title('Input (signed log10)')
axs[0, 1].imshow(tgt_log, cmap='viridis')
axs[0, 1].set_title('Target (signed log10)')
axs[0, 2].imshow(pred_log, cmap='viridis')
axs[0, 2].set_title('Prediction (signed log10)')

# second row: errors
im0 = axs[1, 0].imshow(abs_error, cmap='magma')
axs[1, 0].set_title('Absolute Error')
fig.colorbar(im0, ax=axs[1, 0], fraction=0.046, pad=0.04)

im1 = axs[1, 1].imshow(squared_error, cmap='magma')
axs[1, 1].set_title('Squared Error')
fig.colorbar(im1, ax=axs[1, 1], fraction=0.046, pad=0.04)

# residual (prediction - target) in log space for structural discrepancy
residual_log = signed_log10(pred - tgt).numpy()
im2 = axs[1, 2].imshow(residual_log, cmap='bwr', vmin=-np.max(np.abs(residual_log)), vmax=np.max(np.abs(residual_log)))
axs[1, 2].set_title('Residual (signed log10)')
fig.colorbar(im2, ax=axs[1, 2], fraction=0.046, pad=0.04)

for ax in axs.flatten():
    ax.axis('off')

plt.suptitle(f'GAN Prediction Slice z={depth_idx} | MAE={mae:.4e}, RMSE={rmse:.4e}', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.tensorboard import SummaryWriter
import torch.nn.utils as tnn_utils
import time

# ==== 1. 배치/디바이스 정리 유틸 ====
def prepare_batch(x, y, model):
    """
    채널 차원 보정, dtype/디바이스 일치시키는 함수.
    model의 device를 기준으로 입력을 맞춘다.
    """
    device = next(model.parameters()).device
    if x.ndim == 4:  # (B, D, H, W) -> (B,1,D,H,W)
        x = x[:, None]
    if y.ndim == 4:
        y = y[:, None]
    x = x.float().to(device)
    y = y.float().to(device)
    return x, y

# ==== 2. TensorBoard writer 초기화 ====
writer = SummaryWriter(log_dir="runs/gan_debug")  # 원하는 경로로 바꿔도 됨

# ==== 3. 손실/로그 기록 함수 ====
def log_step(step, loss_components, optimizer_G, optimizer_D, G, D, inputs, targets, fake):
    device = next(G.parameters()).device
    # Loss breakdown
    writer.add_scalar("loss/D_LSGAN", loss_components["loss_D_LSGAN"].item(), step)
    writer.add_scalar("loss/G_LSGAN", loss_components["loss_G_LSGAN"].item(), step)
    writer.add_scalar("loss/FeatureMatching", loss_components["loss_FM"].item(), step)
    writer.add_scalar("loss/Correlation", loss_components["loss_CC"].item(), step)
    writer.add_scalar("loss/G_total", loss_components["loss_G_total"].item(), step)

    # Learning rates
    writer.add_scalar("lr/Generator", optimizer_G.param_groups[0]["lr"], step)
    writer.add_scalar("lr/Discriminator", optimizer_D.param_groups[0]["lr"], step)

    # Gradient norms
    grad_norm_G = 0.0
    for p in G.parameters():
        if p.grad is not None:
            grad_norm_G += p.grad.norm().item() ** 2
    grad_norm_G = grad_norm_G ** 0.5
    writer.add_scalar("grad_norm/Generator", grad_norm_G, step)

    grad_norm_D = 0.0
    for p in D.parameters():
        if p.grad is not None:
            grad_norm_D += p.grad.norm().item() ** 2
    grad_norm_D = grad_norm_D ** 0.5
    writer.add_scalar("grad_norm/Discriminator", grad_norm_D, step)

    # 중간 슬라이스 이미지 (batch 0)
    depth_idx = inputs.shape[2] // 2
    def normalize_for_image(tensor):
        img = tensor[0, 0, depth_idx].detach().cpu()
        img = (img - img.min()) / (img.max() - img.min() + 1e-6)
        return img.unsqueeze(0)  # 1xHxW

    writer.add_image("slice/Input", normalize_for_image(inputs), step)
    writer.add_image("slice/Target", normalize_for_image(targets), step)
    writer.add_image("slice/Fake", normalize_for_image(fake), step)
    writer.add_image("slice/Residual", normalize_for_image(fake - targets), step)

# ==== 4. 예시: training step 안에서 쓰는 패턴 ====
# (전제: G, D, opt, criterion=Pix2PixCCLoss3D(opt), optimizer_G, optimizer_D,
#  scheduler_G, scheduler_D, loader, device가 정의되어 있음)

# 빠른 디버그용: 처음 N개 배치만
max_batches = 5
global_step = 0
G.train(); D.train()

for epoch in range(1):  # 한 epoch만
    for i, (raw_x, raw_y) in enumerate(loader):
        if i >= max_batches:
            break  # 5개 배치 후 종료

        x, y = prepare_batch(raw_x, raw_y, G)  # 디바이스/차원 정리

        components = compute_gan_loss_components(D, G, x, y, opt)
        loss_G = components["loss_G_total"]
        loss_D = components["loss_D_LSGAN"]

        optimizer_G.zero_grad()
        loss_G.backward()
        torch.nn.utils.clip_grad_norm_(G.parameters(), max_norm=5.0)
        optimizer_G.step()

        optimizer_D.zero_grad()
        loss_D.backward()
        torch.nn.utils.clip_grad_norm_(D.parameters(), max_norm=5.0)
        optimizer_D.step()

        # 로그 (예: TensorBoard)
        log_step(global_step, components, optimizer_G, optimizer_D, G, D, x, y, components["fake"])

        print(f"[Step {global_step}] G_loss={loss_G.item():.4f} D_loss={loss_D.item():.4f}")
        global_step += 1

    # scheduler는 에폭 끝에서 한 번만
    scheduler_G.step()
    scheduler_D.step()

```

```python
import matplotlib.pyplot as plt

def plot_gan_diagnostics(df, smooth_window=5):
    """
    1x3 플롯:
      [0] Generator vs Discriminator loss
      [1] Learning rates for G and D
      [2] Gradient norms for G and D
    df: DataFrame with columns 'step','loss_G','loss_D','lr_G','lr_D','grad_norm_G','grad_norm_D'
    smooth_window: moving average window for smoothing noisy curves
    """
    def smooth(x):
        return x.rolling(smooth_window, min_periods=1, center=True).mean()

    fig, axs = plt.subplots(1, 3, figsize=(18, 4), constrained_layout=True)

    # Losses
    axs[0].plot(df['step'], smooth(df['loss_G']), label='Generator Loss')
    axs[0].plot(df['step'], smooth(df['loss_D']), label='Discriminator Loss')
    axs[0].set_title("Loss (smoothed)")
    axs[0].set_xlabel("Step")
    axs[0].set_ylabel("Loss")
    axs[0].legend()
    axs[0].grid(True)

    # Learning rates
    axs[1].plot(df['step'], df['lr_G'], label='LR Generator')
    axs[1].plot(df['step'], df['lr_D'], label='LR Discriminator')
    axs[1].set_title("Learning Rates")
    axs[1].set_xlabel("Step")
    axs[1].set_ylabel("LR")
    axs[1].legend()
    axs[1].grid(True)

    # Gradient norms
    axs[2].plot(df['step'], smooth(df['grad_norm_G']), label='GradNorm G')
    axs[2].plot(df['step'], smooth(df['grad_norm_D']), label='GradNorm D')
    axs[2].set_title("Gradient Norms (smoothed)")
    axs[2].set_xlabel("Step")
    axs[2].set_ylabel("Norm")
    axs[2].legend()
    axs[2].grid(True)

    plt.suptitle("GAN Training Diagnostics", fontsize=16)
    plt.show()
```
